<h1 align='center'>Proyecto final sobre librerias Panda, Bokeh y Jupyter</h1>

<p>Primero se importan las librearias necesarias, se lee el archivo csv y se presenta el dataframe</p>

```python
#Se importan las librerias y sus objetos
import pandas as pd
import numpy as np
from bokeh.io import show, output_file, output_notebook
from bokeh.plotting import figure
from bokeh.transform import cumsum
from bokeh.models import ColumnDataSource
from bokeh.transform import factor_cmap
from math import pi
output_notebook()

#Se lee al archivo y se presenta el dataframe
datos = pd.read_csv('data/unidadesdesaludconvertcsv-csv-oprf.csv')
datos
```

<p>Se cuentan cuantos cetros por provincias existen y se almacena los resultados en num_centros</p>

```python
num_centros = datos.groupby(['Provincia'])['Provincia'].count()
```

<h4>Figura 1</h4>
<p>En esta figura de pastel, se da a conocer cuantos centros de salud existen por provincia en el Ecuador</p>

```python
from bokeh.palettes import inferno #Se importa la paleta de


output_file("pie_centros.html")# Se crea el archivo de salida

x = num_centros # La variable x toma el valor de la cantidad de centros antes sacado

# Se crea una serie de pandas, se resetea el valor de los indices y se asigna la cabecera a aprovincia
data = pd.Series(x).reset_index(name='value').rename(columns={'index':'Provincia'}) 
data['angle'] = data['value']/data['value'].sum() * 2*pi # Se calcula el angulo que van a tener los datos dentro del circulo
data['color'] = inferno(24) # Se asignan los colores

# Se crea la figura, la variable tooltips permite ver al usuario los datos la pasar el ratos por el grafico
p = figure(plot_height=600, plot_width=800,title="Numero de centros por provincia", toolbar_location=None,
           tools="hover", tooltips="@Provincia: @value", x_range=(-0.5, 1.0))

# Se genera el circulo y como leyenda se pone a todas las provincias
p.wedge(x=0, y=1, radius=0.4,
        start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
        line_color="white", fill_color='color', legend='Provincia', source=data)

p.axis.axis_label=None
p.axis.visible=False
p.grid.grid_line_color = None
# se presenta la grafica
show(p)
```

<p>Se cuentan la cantidad de areas que exixte para cada una y se agregan las areas sin que se repitan a areas2</p>

```python
n = datos.groupby(['Area'])['Area'].count()
x = n
# Se crea una serie de pandas, se resetea el valor de los indices y se asigna la cabecera a Area
data = pd.Series(x).reset_index(name='value').rename(columns={'index':'Area'})
data.keys() # Se toman todas las llaves de data
areas2 = data.Area.unique().astype(str) # Se agregan las areas sin se que se repitan de forma de cadena
```

<h4>Figura 2</h4>
<p>En esta figura de barras se presenta la cantidad de areas existentes</p>

```python

# Se importa la paleta de colores
from bokeh.palettes import plasma



output_file("barra_areas.html") #Se crea el archivo de salida

areas = areas2 # areas toma el valor de las areas en forma de cadena
counts = n # counts toma el valor de la cantidad de areas

source = ColumnDataSource(data=dict(areas=areas, counts=counts)) # Se almacenan los datos de la grafica

# Se crea la figura
p = figure(x_range=areas, plot_height=400, plot_width = 900,toolbar_location=None, title="Numero de Areas")
# Se generan las barras
p.vbar(x='areas', top='counts', width=0.9, source=source,
       line_color='white', fill_color=factor_cmap('areas', palette=plasma(34), factors=areas))

p.xgrid.grid_line_color = None
p.y_range.start = 0 # EL valor desde donde epieza el grafico en el eje y
p.y_range.end = 430 # EL valor hasta donde llegua el grafico en el eje y
# Se crean las etiquetas de el eje x y el eje y
p.yaxis.axis_label = "Numero De Areas"
p.xaxis.axis_label = "Area"


show(p)
```

<h4>Figura 3</h4>
<p>En esta figura de barras se muestra la cobertura que tienen los centros de atencion</p>

```python
# Se cuenta la cantidad de coberturas que pfrecen los centros
n_cobertura = datos.groupby(['Cobertura'])['Cobertura'].count().tolist()
```

```python
from bokeh.palettes import Spectral6 # Se importa la paleta de colores


output_file("barra_coberturas.html") # Se crea el archivo de salida
# Se crea la lista de coberturas disponibles y el valor de cada una
cobertura = ['No Registrada', 'Cantonal', 'Provincial']
counts = [5, n_cobertura[5], n_cobertura[4]]

source = ColumnDataSource(data=dict(cobertura=cobertura, counts=counts)) # Se almacenen los datos de la grafica

# Se cre la figura
p = figure(x_range=cobertura, plot_height=400, plot_width = 900,toolbar_location=None, title="Cobertura de centros de salud")
# Se generan las barras
p.vbar(x='cobertura', top='counts', width=0.9,source=source, legend='cobertura',
       line_color='#000000', fill_color=factor_cmap('cobertura', palette=Spectral6, factors=cobertura))

p.xgrid.grid_line_color = None
p.y_range.start = 0 # EL valor desde donde epieza el grafico en el eje y
p.y_range.end = 180 # El valor hasta donde llega el grafico en el eje y
# Ubicacion de la leyenda
p.legend.orientation = "horizontal"
p.legend.location = "top_right"
# Etiquetas del graafico para amnos ejes
p.yaxis.axis_label = "Total"
p.xaxis.axis_label = "Coberturas"


show(p)# Se presenta el grafico
```

<h4>Figura 4</h4>
<p>En esta figura se muetran los tipos de centro de atencion que existen a nivel nacional</p>

```python
# Se cuenta cuanatos datos de la llave tipo_atencion existen
t_atencion = datos.groupby(['Tipo_atencion'])['Tipo_atencion'].count()
```

```python
from bokeh.palettes import magma # Se importa la paleta de colores


output_file("pie_atencion.html") # se crea el archivo de salida

x = t_atencion # x toma el valor de la cantidad de tipos de atencion existentes 

# Se crea una serie de pandas, se resetea el valor de los indices y se asigna la cabecera a Tipo_atencion
data = pd.Series(x).reset_index(name='value').rename(columns={'index':'Tipo_atencion'})
data['angle'] = data['value']/data['value'].sum() * 2*pi # Se calcula el valor de los angulos que van a tener los datos en el circulo
data['color'] = magma(33) # Se agrega el color

# Se crea la figura, la variable tooltips permite ver al usuario los datos la pasar el ratos por el grafico
p = figure(plot_height=1000, plot_width=800,title="Cantidad de centros de atencion a nivel nacional por su tipo de atencion", toolbar_location=None,
           tools="hover", tooltips="@Tipo_atencion: @value", x_range=(-0.5, 1.0))

# Se genera el circulo y como leyenda se pone a todas los tipos de atencion que brindan los centrs
p.wedge(x=0, y=1, radius=0.3,
        start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
        line_color="white", fill_color='color', legend='Tipo_atencion', source=data)

p.axis.axis_label=None
p.axis.visible=False
p.grid.grid_line_color = None

show(p) # se presenta la grafica
```
